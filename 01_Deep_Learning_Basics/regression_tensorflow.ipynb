```python
# regression_tensorflow.ipynb

# ğŸ“˜ Linear Regression with TensorFlow 2.x 

# Objective: Predict 'y' given 'x' using a linear relationship (y = mx + b)

# ğŸ“¦ Step 1: Import Libraries
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# âœ… Check TensorFlow version
print("TensorFlow version:", tf.__version__)

# ğŸ“Š Step 2: Create Dummy Dataset
# y = 3x + 7 + noise

np.random.seed(42)
X = np.linspace(0, 10, 100)
noise = np.random.normal(0, 1, X.shape)
Y = 3 * X + 7 + noise

# Visualize data
plt.scatter(X, Y)
plt.title("Generated Data: y = 3x + 7 + noise")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()

# ğŸ“¦ Step 3: Build Linear Regression Model using tf.keras
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=1, input_shape=[1])
])

# ğŸ“‰ Step 4: Compile Model
model.compile(optimizer='sgd', loss='mean_squared_error')

# ğŸ” Step 5: Train the Model
history = model.fit(X, Y, epochs=200, verbose=0)

# ğŸ“ˆ Plot Loss Curve
plt.plot(history.history['loss'])
plt.title("Loss Curve")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.show()

# ğŸ§  Step 6: View Learned Parameters (Weight and Bias)
weights, biases = model.layers[0].get_weights()
print(f"Learned Weight (m): {weights[0][0]:.2f}")
print(f"Learned Bias (b): {biases[0]:.2f}")

# ğŸ” Step 7: Make Predictions
Y_pred = model.predict(X)

# ğŸ“Š Step 8: Visualize Predictions
plt.scatter(X, Y, label='Original Data')
plt.plot(X, Y_pred, color='red', label='Model Prediction')
plt.title("Linear Regression Fit")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()

# ğŸ¯ Done!
print("âœ… Linear regression completed successfully.")
```
